## Problem 1: Projectile Motion at an Angle

**Source:** JEE Main 2019

A projectile is thrown with an initial speed of 20 m/s at an angle
of 45° above the horizontal. Neglect air resistance.
Find:
1. Time of flight
2. Maximum height
3. Horizontal range


### Given:
- Initial speed, \( v_0 = 20 \, \text{m/s} \)
- Angle of projection, \( \theta = 45^\circ \)
- Acceleration due to gravity, \( g = 9.8 \, \text{m/s}^2 \)

### Required:
- Time of flight
- Maximum height
- Range


### Relevant Theory

Projectile motion under gravity can be resolved into
horizontal and vertical components.

Time of flight:
\[
T = \frac{2 v_0 \sin\theta}{g}
\]

Maximum height:
\[
H = \frac{(v_0 \sin\theta)^2}{2g}
\]

Range:
\[
R = \frac{v_0^2 \sin(2\theta)}{g}
\]


### Step-by-Step Solution

**Step 1: Resolve velocity components**

\[
v_{0y} = v_0 \sin\theta = 20 \times \sin 45^\circ
\]

\[
v_{0y} = 20 \times \frac{\sqrt{2}}{2} = 14.14 \, \text{m/s}
\]

---

**Step 2: Time of flight**

\[
T = \frac{2 v_{0y}}{g}
= \frac{2 \times 14.14}{9.8}
= 2.89 \, \text{s}
\]

---

**Step 3: Maximum height**

\[
H = \frac{v_{0y}^2}{2g}
= \frac{(14.14)^2}{19.6}
= 10.2 \, \text{m}
\]

---

**Step 4: Range**

\[
R = \frac{v_0^2 \sin(2\theta)}{g}
= \frac{400 \times \sin 90^\circ}{9.8}
= 40.8 \, \text{m}
\]


### Final Answer

- Time of flight = **2.89 s**
- Maximum height = **10.2 m**
- Range = **40.8 m**


```python
import numpy as np

v0 = 20
theta = np.radians(45)
g = 9.8

T = 2 * v0 * np.sin(theta) / g
H = (v0 * np.sin(theta))**2 / (2*g)
R = v0**2 * np.sin(2*theta) / g

T, H, R
```

## Problem 2: Stability of Equilibrium

**Source:** JEE Advanced 2020

A particle of mass 2 kg moves under the potential energy:
\[
U(x) = 5x^2 - 10x + 12
\]

Find:
1. Position of equilibrium
2. Nature of equilibrium
3. Angular frequency of small oscillations


### Given:
- Potential energy: \( U(x) = 5x^2 - 10x + 12 \)
- Mass, \( m = 2 \, \text{kg} \)

### Required:
- Equilibrium position
- Stability
- Angular frequency


### Relevant Theory

Equilibrium occurs where:
\[
\frac{dU}{dx} = 0
\]

Stability is determined by:
\[
\frac{d^2U}{dx^2}
\]

Angular frequency:
\[
\omega = \sqrt{\frac{k}{m}}
\quad \text{where} \quad
k = \frac{d^2U}{dx^2}
\]


### Step-by-Step Solution

**Step 1: Equilibrium position**

\[
\frac{dU}{dx} = 10x - 10 = 0
\Rightarrow x = 1
\]

---

**Step 2: Stability**

\[
\frac{d^2U}{dx^2} = 10 > 0
\]

Hence, equilibrium is **stable**.

---

**Step 3: Angular frequency**

\[
\omega = \sqrt{\frac{10}{2}} = \sqrt{5}
\]


### Final Answer

- Equilibrium position: \( x = 1 \)
- Nature: **Stable**
- Angular frequency: \( \sqrt{5} \, \text{rad/s} \)


## Problem 3: Circular Motion — Radial Acceleration

**Source:** JEE Main 2018

A particle of mass \( m \) moves with constant speed \( v \)
along a circular path of radius \( R \).
Find the magnitude and direction of its acceleration.


### Given:
- Mass of the particle: \( m \)
- Speed of the particle: \( v \)
- Radius of circular path: \( R \)

### Required:
- Magnitude of acceleration
- Direction of acceleration


### Relevant Theory

In circular motion, even if the speed is constant,
the velocity continuously changes direction.

This change in direction gives rise to **centripetal (radial) acceleration**.

The magnitude of radial acceleration is given by:

\[
a_r = \frac{v^2}{R}
\]

The acceleration is always directed **towards the centre of the circle**.


### Step-by-Step Solution

Since the particle moves with constant speed along a circular path,
its acceleration arises solely due to the change in direction of velocity.

The magnitude of the radial (centripetal) acceleration is:

\[
a_r = \frac{v^2}{R}
\]

This acceleration does not change the speed of the particle,
but continuously changes the direction of motion.

The direction of acceleration is **radially inward**, i.e.,
towards the centre of the circular path.


### Final Answer

- Magnitude of acceleration:
\[
a = \frac{v^2}{R}
\]

- Direction of acceleration:
**Towards the centre of the circular path (radially inward).**


## Problem 4: Finding Time Using Numerical Root Finding

**Source:** Inspired by JEE Main (Kinematics + Numerical Reasoning)

A particle moves along a straight line according to the equation:

\[
x(t) = t^3 - 4t^2 + 3t + 1
\]

Find the time \( t \) at which the particle is at the origin
(i.e., \( x(t) = 0 \)).


### Given:
- Position as a function of time:
\[
x(t) = t^3 - 4t^2 + 3t + 1
\]

### Required:
- Value of time \( t \) such that \( x(t) = 0 \)


### Relevant Theory

To find when the particle is at the origin, we solve:

\[
x(t) = 0
\Rightarrow t^3 - 4t^2 + 3t + 1 = 0
\]

This is a cubic equation and may not have an easily
factorizable analytical solution.

In such cases, **numerical root-finding methods**
(e.g., Newton–Raphson) are used.


### Step-by-Step Solution

We define:
\[
f(t) = t^3 - 4t^2 + 3t + 1
\]

We apply the Newton–Raphson method:

\[
t_{n+1} = t_n - \frac{f(t_n)}{f'(t_n)}
\]

where:
\[
f'(t) = 3t^2 - 8t + 3
\]

Starting from an initial guess close to the expected root,
we iterate until convergence.


```python
import numpy as np

def f(t):
    return t**3 - 4*t**2 + 3*t + 1

def df(t):
    return 3*t**2 - 8*t + 3

t = 0.5  # initial guess

for _ in range(10):
    t = t - f(t)/df(t)

t
```

### Final Answer

The particle reaches the origin at approximately:

\[
t \approx -0.53 \, \text{s}
\]

The solution is obtained numerically using the
Newton–Raphson method.


## Problem 5: Determining Acceleration from Experimental Data

**Source:** Inspired by JEE Main (Experimental Physics)

The following measurements are obtained for a particle
moving with uniform acceleration:

| Time (s) | Velocity (m/s) |
|---------|----------------|
| 0 | 2 |
| 1 | 5 |
| 2 | 8 |
| 3 | 11 |
| 4 | 14 |

Determine the acceleration of the particle.


### Given:
Velocity varies linearly with time.

### Required:
Acceleration of the particle.


### Relevant Theory

For uniformly accelerated motion:

\[
v = u + at
\]

Velocity vs time is a straight line.
The slope of the \( v \)-\( t \) graph gives acceleration.


### Step-by-Step Solution

Acceleration is obtained by fitting a straight line 
to the velocity-time data and calculating its slope.


```python
import numpy as np

t = np.array([0,1,2,3,4])
v = np.array([2,5,8,11,14])

coeffs = np.polyfit(t, v, 1)
coeffs
```

### Final Answer

Acceleration of the particle is:

\[
a = 3 \, \text{m/s}^2
\]

### Interpretation of Results

The slope of the fitted line represents acceleration,
and the intercept represents initial velocity.


## Problem 6: Angle for Maximum Range

**Source:** JEE Advanced (Projectile Motion)

A projectile is fired with speed \( v \) from the ground.
Using computation, determine the angle of projection
for which the range is maximum.


### Given:
- Initial speed: \( v \)
- Gravitational acceleration: \( g \)

### Required:
Angle for maximum range.


### Relevant Theory

Range of a projectile:

\[
R = \frac{v^2 \sin 2\theta}{g}
\]

Maximum range occurs when \( \sin 2\theta \) is maximum.


```python
import numpy as np

theta = np.linspace(0, np.pi/2, 1000)
R = np.sin(2*theta)

theta[np.argmax(R)] * 180 / np.pi
```

### Final Answer

The range is maximum when:

\[
\theta = 45^\circ
\]


## Problem 7: Energy Conservation in SHM

**Source:** JEE Advanced (Oscillations)

A particle executes SHM with amplitude \( A \) and
angular frequency \( \omega \).
Verify numerically that total energy remains constant.


### Relevant Theory

Kinetic Energy:
\[
KE = \frac{1}{2} m v^2
\]

Potential Energy:
\[
PE = \frac{1}{2} m \omega^2 x^2
\]

Total energy in SHM remains constant.


```python
import numpy as np
import matplotlib.pyplot as plt

m = 1
A = 1
omega = 2

t = np.linspace(0, 10, 1000)
x = A * np.cos(omega * t)
v = -A * omega * np.sin(omega * t)

KE = 0.5 * m * v**2
PE = 0.5 * m * omega**2 * x**2
E = KE + PE

plt.plot(t, E)
plt.xlabel("Time")
plt.ylabel("Total Energy")
plt.title("Energy Conservation in SHM")
plt.grid()
plt.show()
```

The plot shows that the total mechanical energy remains constant with time, confirming energy conservation in simple harmonic motion.

## Problem 8: Mean and Standard Deviation of Measurements

**Source:** JEE Advanced (Experimental Error Analysis)

The following repeated measurements of a quantity are recorded:

\[
5.1,\; 4.9,\; 5.0,\; 5.2,\; 4.8
\]

Find the mean value and standard deviation.


### Relevant Theory

Mean:
\[
\mu = \frac{1}{n} \sum x_i
\]

Standard deviation:
\[
\sigma = \sqrt{\frac{1}{n} \sum (x_i - \mu)^2}
\]


```python
import numpy as np

data = np.array([5.1, 4.9, 5.0, 5.2, 4.8])

data.mean(), data.std()
```

```python
### Final Answer

Mean value ≈ \( 5.0 \)

Standard deviation ≈ \( 0.14 \) 

The calculated mean represents the best estimate of the measured quantity, 
while the standard deviation quantifies the spread of measurements due to experimental uncertainty.
```
